```python
# Print the number of words starting with consonants and the number of words starting with vowels in
# all messages.

n = len(df)
row = len(df.loc[0])
vowels = []
for i in range(n):
    msg = df.loc[i][1]
    msg = msg.split()
    for word in msg:
        r = re.findall(r'^[AEIOUaeiou][^^]*',word)
        vowels.extend(r)
print(len(vowels))
print(vowels)
```

```python
n = len(df)
row = len(df.loc[0])
consonants = []
for i in range(n):
    msg = df.loc[i][1]
    msg = msg.split()
    for word in msg:
        r = re.findall(r'^[B-DF-HJ-NP-TV-Zb-df-hj-np-tv-z][^^]*',word)
        consonants.extend(r)
print(len(consonants))
print(consonants)
```

```python
# Compute the percentage of capitalized words in spam and ham messages.

n = len(df)
row = len(df.loc[0])
capitalized_spam = []
capitalized_ham = []
spam = []
ham = []

for i in range(n):
#     bbb=0
    s_or_h = df.loc[i][0]
    if s_or_h == "spam":
        msg = df.loc[i][1]
        msg = msg.split()
        for word in msg:
            spam.append(word)
            r=re.findall(r'^[^(a-z^)]*[A-Z!@#$%&*()_+~`.,/;\]{}[><?:"-=0-9]*[A-Z]+[A-Z!@#$%&*()_+~`.,/;\]{}[><?:"-=0-9]*[^(a-z^)]*$',word)
            capitalized_spam.extend(r)
#             if r:
#                 bbb=bbb+1
#         print(bbb)
   
            
    if s_or_h == "ham":
        msg = df.loc[i][1]
        msg = msg.split()
        for word in msg:
            ham.append(word)
            r=re.findall(r'^[^(a-z^)]*[A-Z!@#$%&*()_+~`.,/;\]{}[><?:"-=0-9]*[A-Z]+[A-Z!@#$%&*()_+~`.,/;\]{}[><?:"-=0-9]*[^(a-z^)]*$',word)
            capitalized_ham.extend(r)


print("Total words in ham : ", len(ham))
# print(capitalized_ham)
print("capitalized words in ham : ",len(capitalized_ham))
ratio_ham = len(capitalized_ham)/len(ham)
print("capitalized words % in ham : ",ratio_ham*100)
print()

print("Total words in spam : ", len(spam))
# print(capitalized_spam)
print("capitalized words in spam : ",len(capitalized_spam))
ratio_spam = len(capitalized_spam)/len(spam)
print("capitalized words % in spam : ",ratio_spam*100)
print()

cspam=ratio_spam*100

tm = (len(spam)+len(ham))/(len(capitalized_ham)+len(capitalized_spam))
print("Total messages in ham and spam :",len(spam)+len(ham))
print("Total Capitalized words in ham and spam :", len(capitalized_ham)+len(capitalized_spam))
print("Overall Capitalized % ",(tm**-1)*100)
```

```python
# List all email ids and phone/mobile numbers and their respective counts in the dataset. Compute
# percentage for spam and ham messages for both email and mobile numbers.

n = len(df)
row = len(df.loc[0])
phone_spam = []
phone_ham = []
spam = []
ham = []
for i in range(n):
    s_or_h = df.loc[i][0]
    if s_or_h == "spam":
        msg = df.loc[i][1]
        msg = msg.split()
        for word in msg:
            spam.append(word)
            r=re.findall(r'[0-9]?[0-9]{10}|[(]?[0-9]{4,5}[)]?-?[(]?[0-9]{3,4}[)]?-?[(]?[0-9]{4,5}[)]?',word)
            phone_spam.extend(r)
    if s_or_h == "ham":
        msg = df.loc[i][1]
        msg = msg.split()
        for word in msg:
            ham.append(word)
            r=re.findall(r'[0-9]?[0-9]{10}|[(]?[0-9]{4,5}[)]?-?[(]?[0-9]{3,4}[)]?-?[(]?[0-9]{4,5}[)]?',word)
            phone_ham.extend(r)
            

print(phone_spam)
print(phone_ham)
print(len(phone_spam))
print(len(phone_ham))
print()

print("Total words in ham : ", len(ham))
# print(phone_ham)
print("phone in ham : ",len(phone_ham))
ratio_ham = len(phone_ham)/len(ham)
print("phone % in ham : ",ratio_ham*100)
print()

print("Total words in spam : ", len(spam))
# print(phone_spam)
print("phone in spam : ",len(phone_spam))
ratio_spam = len(phone_spam)/len(spam)
print("phone % in spam : ",ratio_spam*100)
print()

tm = (len(spam)+len(ham))/(len(phone_ham)+len(phone_spam))
print("Total phones in ham and spam :",len(spam)+len(ham))
print("Total phones in ham and spam :", len(phone_ham)+len(phone_spam))
print("Overall phones % ",(tm**-1)*100)
```

```python
# List all monetary quantities, e.g., £1000. Compute the percentage of spam and ham messages with
# monetary quantities.

n = len(df)
row = len(df.loc[0])
money_spam = []
money_ham = []
spam = []
ham = []
for i in range(n):
    s_or_h = df.loc[i][0]
    if s_or_h == "spam":
        msg = df.loc[i][1]
        msg = msg.split()
        for word in msg:
            spam.append(word)
            r=re.findall(r'[$£][0-9]+[.,]?[0-9]*[,]?[0-9]*',word)
            money_spam.extend(r)
    if s_or_h == "ham":
        msg = df.loc[i][1]
        msg = msg.split()
        for word in msg:
            ham.append(word)
            r=re.findall(r'[$£][0-9]+[.,]?[0-9]*[,]?[0-9]*',word)
            money_ham.extend(r)
            

print(money_spam)
print(money_ham)
print(len(money_spam))
print(len(money_ham))


print("Total words in ham : ", len(ham))
print("monetary words in ham : ",len(money_ham))
ratio_ham = len(money_ham)/len(ham)
print("monetary words % in ham : ",ratio_ham*100)
print()

print("Total words in spam : ", len(spam))
print("monetary words in spam : ",len(money_spam))
ratio_spam = len(money_spam)/len(spam)
print("monetary words % in spam : ",ratio_spam*100)
print()

mspam=ratio_spam*100

tm = (len(spam)+len(ham))/(len(money_ham)+len(money_spam))
print("Total messages in ham and spam :",len(spam)+len(ham))
print("Total monetary words in ham and spam :", len(money_ham)+len(money_spam))
print("Overall monetary % ",(tm**-1)*100)

```

```python
# Count and print all emoticons (use NLTK only) in the dataset.

n = len(df)
row = len(df.loc[0])
emoticons = []
for i in range(n):
    msg = df.loc[i][1]
    tk = RegexpTokenizer('[@*]?[|:;=X^]+[-''=*]?[DPp0*Oo>^)/\(@]+',gaps=False)
    tokens = tk.tokenize(msg)
    for i in tokens:
        emoticons.append(i)
#     for word in msg:
#         r = re.findall(r'[@*]?[|:;=X^]+[-''=*]?[DPp0*Oo>^)/\(@]+',word)
#         emoticons.extend(r)
print(len(emoticons))
print(emoticons)
```

```python
# Print a list of words with clitics (a subword unit that cannot stand on it own, e.g., ‘ve, n’t, ‘d)

n = len(df)
row = len(df.loc[0])
clitics = []
for i in range(n):
    msg = df.loc[i][1]
    msg = msg.split()
    for word in msg:
        r = re.findall(r"\w+[‘''’]\w+",word)
        clitics.extend(r)
        
print(len(clitics))
print(clitics)

```
